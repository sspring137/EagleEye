## 2. EagleEye_v7

Reference dataset: $\mathbf{X}$

Test dataset:      $\mathbf{Y}$ 

Define: $\mathbf{U} = \mathbf{X} \cup \mathbf{Y}$

COMPUTE: $\forall i\in U,$  $\Upsilon_i $ whith Inject validation for the data belonging to $\mathbf{X}$ 

Revert the order of Reference and Test to get $\forall j\in U,$  $\Upsilon_{j}(rev) $ whith Inject validation for the data belonging to $\mathbf{Y}$ 

```python
K_M                   = 400
CRITICAL_QUANTILES    = [ 1-1E-4, 1-1E-5 ]
VALIDATION            = reference_data.shape[0]


NUM_CORES             = 10

result_dictionary     = EagleEye_v7.Soar(
    reference_data, test_data, result_dict_in = {}, K_M = K_M, critical_quantiles = CRITICAL_QUANTILES,
    num_cores=NUM_CORES, validation=VALIDATION, partition_size=100 )

```

#### Overdensities test_set:
result_dictionary['stats']['Upsilon_i_plus'] : $\Upsilon_i, i\in \mathbf{Y} $ 

#### Underdensities reference_set:
result_dictionary['stats']['Upsilon_i_Val_plus'] : $\Upsilon_i, i\in \mathbf{X} $ (the IV one)

#### Overdensities reference_set:
result_dictionary['stats']['Upsilon_i_plus'] : $\Upsilon_j(rev), j\in \mathbf{X} $ 

#### Underdensities test_set:
result_dictionary['stats']['Upsilon_i_Val_plus'] : $\Upsilon_j(rev), j\in \mathbf{Y} $ (the IV one)

## 3 Iterative Equalization

Pipeline for
#### Overdensities test_set:
result_dictionary['stats']['Upsilon_i_plus'] : $\Upsilon_i, i\in \mathbf{Y}$; 

1) While $\exists i \in \mathbf{Y}$ such that $\Upsilon_i \geq \Upsilon_*$;

2) Identify the index $i_{Max}$ such that maximize $\Upsilon_{i_{Max}} = \Max_{i\in \mathbf{Y}}{\Upsilon_i}$;

\begin{enumerate}
    \item \textbf{While} there exists a point $i$ in $\mathbf{Y}$ whose statistic $\Upsilon_i \geq \Upsilon_*$:
    \begin{enumerate}
        \item Identify the index $i_{\text{Max}}$ of the point in $\mathbf{Y}$ with the highest $\Upsilon_i$.
        \item Remove $i_{\text{Max}}$ and its nearest neighbours up to the next $\mathbf{X}$ data point from the combined set $\mathbf{U} = \mathbf{X} \cup \mathbf{Y}$. Store all removed points in a new set $\mathbf{W_o}$.
        \item Recompute $\Upsilon_i$ for the remaining points in $\mathbf{Y}$ that still exceed $\Upsilon_*$, restricting the nearest-neighbour search to $\mathbf{U} \setminus \mathbf{W_o}$.
        \item Repeat until no points in $\mathbf{Y}$ exceed $\Upsilon_*$.
    \end{enumerate}
    \item \textbf{Reverse roles of $\mathbf{X}$ and $\mathbf{Y}$} to capture \emph{overdensity} anomalies in $\mathbf{X}$ that induce \emph{underdensities} in the test set. Repeating the above procedure yields a second set $\mathbf{W_u}$, containing those portions of $\mathbf{X}$ that drive local underdensity imbalances in $\mathbf{Y}$.

\end{enumerate}
